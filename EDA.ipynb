```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('Data_cleaned.csv')
data.head()
```

```python
# Display summary statistics for numerical columns
numerical_summary = data.describe()
print(numerical_summary)
```

#### 1. Distribution of hours of sleep

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set up visualization style
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)

# Create a dictionary to store insights
eda_insights = {}

# 1. Distribution of hours of sleep
plt.figure()
sns.histplot(data['hours of sleep had last night'], bins=20, kde=True, color='blue')
plt.title("Distribution of Hours of Sleep")
plt.xlabel("Hours of Sleep")
plt.ylabel("Frequency")
plt.grid(axis='y')
eda_insights['sleep_distribution'] = "Most children had 9-11 hours of sleep, with few outliers sleeping very little or a lot."
plt.show()


```

#### 2. Relationship between brushing teeth and hours of sleep

```python
# 2. Relationship between brushing teeth and hours of sleep
plt.figure()
sns.boxplot(x=data['3. How many times did you brush your teeth YESTERDAY?'],
            y=data['hours of sleep had last night'], palette='Set3')
plt.title("Brushing Teeth Frequency vs Hours of Sleep")
plt.xlabel("Brushing Teeth Frequency (Yesterday)")
plt.ylabel("Hours of Sleep")
eda_insights['brushing_vs_sleep'] = "Children who brushed their teeth more frequently tend to have consistent sleep hours."
plt.show()
```

#### 3. Correlation heatmap for numerical features

```python
# 3. Correlation heatmap for numerical features
numerical_features = data.select_dtypes(include=['number'])
plt.figure()
sns.heatmap(numerical_features.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap of Numerical Features")
eda_insights['correlation'] = "Strong positive correlations observed between general wellbeing factors (health, school, family, friends). Sleep hours showed weak correlation with other factors."
plt.show()

```

#### 4. Boxplot for general wellbeing factors

```python
# 4. Boxplot for general wellbeing factors
wellbeing_factors = ['Your Health ', 'Your School', 'Your Family', 'Your Friends', 
                     'Your Appearance (how you look)', 'Your Life']
plt.figure()
data[wellbeing_factors].boxplot()
plt.title("Boxplot of General Wellbeing Factors")
plt.ylabel("Score (0-10)")
eda_insights['wellbeing_boxplot'] = "Most wellbeing scores are consistently high (above 7), indicating children feel positive overall."
plt.show()

# Document insights for further analysis
eda_insights
```

#### Key Insights from Exploratory Data Analysis (EDA):

##### Distribution of Hours of Sleep:

* Most children slept between 9 to 11 hours, which aligns with recommended sleep for their age group.

* A few outliers slept significantly less or more, indicating potential irregularities or unique cases.

##### Brushing Teeth vs. Sleep Hours:

* Children who brushed their teeth more frequently tend to have more consistent and possibly healthier sleep patterns.
There seems to be no major anomalies in brushing habits affecting sleep.

##### Correlation Analysis:

* Strong positive correlations exist between general wellbeing factors (e.g., health, school, family, friends).
Sleep hours showed a weak correlation with other factors, suggesting it is influenced by independent variables.

##### Wellbeing Scores:

* Most wellbeing scores are consistently high (above 7 on a scale of 10), indicating that children generally feel positive about their health, school, family, and life.

##### Visualizations:

* Histogram: Shows the distribution of hours of sleep.

* Boxplot: Explores the relationship between brushing frequency and sleep hours.

* Heatmap: Displays correlations between numerical features, highlighting key relationships.

* Wellbeing Boxplot: Summarizes the distribution of wellbeing scores across various dimensions.

####  EDA on emotional difficulty columns

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Select emotional difficulty columns
emotional_difficulty_columns = [
    "24. Remember, there are no right or wrong answers, just pick which is right for you. [I feel lonely]",
    "24. Remember, there are no right or wrong answers, just pick which is right for you. [I cry a lot]",
    "24. Remember, there are no right or wrong answers, just pick which is right for you. [I worry a lot]",
    "24. Remember, there are no right or wrong answers, just pick which is right for you. [I feel nobody likes me]",
    "24. Remember, there are no right or wrong answers, just pick which is right for you. [I am unhappy]"
]

```

```python
# Filter data for emotional difficulty columns
emotional_difficulty_data = data[emotional_difficulty_columns]
```

```python
# Define the mapping for categorical responses to numerical values
response_mapping = {
    "Never": 0,
    "Sometimes": 1,
    "Always": 2
}
```

```python
# Apply the mapping to the emotional difficulty columns
for column in emotional_difficulty_columns:
    emotional_difficulty_data[column] = emotional_difficulty_data[column].map(response_mapping)

```

```python
# Verify if mapping worked
updated_unique_values = {column: emotional_difficulty_data[column].unique() for column in emotional_difficulty_columns}
```

```python
# 1. Boxplot of Emotional Difficulty Indicators
plt.figure(figsize=(12, 6))
sns.boxplot(data=emotional_difficulty_data)
plt.title("Distribution of Emotional Difficulty Indicators (Corrected Categories)")
plt.ylabel("Scale (0 = Never, 2 = Always)")
plt.xticks(rotation=45)
plt.show()

```

```python
# 2. Correlation Heatmap of Emotional Difficulty Indicators
plt.figure(figsize=(10, 8))
sns.heatmap(emotional_difficulty_data.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap of Emotional Difficulty Indicators (Corrected Categories)")
plt.show()
```

```python
# 3. Summary statistics for Emotional Difficulty Indicators
corrected_emotional_difficulty_summary = emotional_difficulty_data.describe()

# Display summary statistics
corrected_emotional_difficulty_summary
```

### Key Insights from Emotional Difficulty Analysis

#### Boxplot of Emotional Difficulty Indicators:

* Most children reported lower levels of emotional difficulties, leaning toward "Never" (0) or "Sometimes" (1).
* Indicators like "I worry a lot" show higher variability and outliers, highlighting significant challenges for some children.

#### Correlation Heatmap:

* "I feel lonely", "I cry a lot", and "I worry a lot" are moderately correlated, suggesting these emotional difficulties often co-occur.
* "I feel nobody likes me" and "I am unhappy" are correlated with loneliness and crying, reinforcing the interconnectedness of these struggles.

#### Summary Statistics:

* "I worry a lot" had the highest mean score (~0.75), indicating a common concern among children.
* "I feel nobody likes me" had the lowest mean score (~0.35), suggesting this is less of an issue but still notable.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Add the "hours of sleep had last night" column to the emotional difficulty dataset
emotional_difficulty_data['hours of sleep had last night'] = data['hours of sleep had last night']
```

```python
# Correlation between emotional difficulties and sleep
correlation_with_sleep = emotional_difficulty_data.corr()['hours of sleep had last night'].drop('hours of sleep had last night')

# Visualize the correlations with sleep
plt.figure(figsize=(10, 6))
correlation_with_sleep.sort_values().plot(kind='barh', color='skyblue')
plt.title("Correlation Between Emotional Difficulties and Sleep Duration")
plt.xlabel("Correlation Coefficient")
plt.ylabel("Emotional Difficulty Indicators")
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()

```

#### Correlation Between Emotional Difficulties and Sleep Duration

##### Insights:

* Most emotional difficulties have a weak negative correlation with sleep duration, suggesting that more sleep slightly reduces emotional challenges.

##### Key Observations:
* "I feel nobody likes me" (-0.10) and "I am unhappy" (-0.08) show the strongest (albeit weak) negative correlations with sleep.
* "I worry a lot" shows a near-zero correlation (+0.0036), suggesting little to no direct impact of sleep on this specific difficulty.

##### Visualization:

* A bar plot was created to visualize the correlation coefficients, making it easier to identify trends.


#### Recommendations Based on Correlation:

##### Promote Adequate Sleep:

* Although the correlations are weak, ensuring children get adequate sleep might help mitigate feelings of unhappiness and loneliness.
* Encourage consistent sleep routines and minimize distractions before bedtime.

##### Address Emotional Difficulties Holistically:

* Since sleep has limited impact on worrying behavior, additional interventions like stress management techniques or counseling may be needed.
